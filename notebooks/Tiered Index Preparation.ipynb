```python
import pandas as pd
import numpy as np
import time
import pickle
```

```python
matrix_test = pickle.load( open( "saved_weghts_matrix.p", "rb" ) )
index= pickle.load( open( "index.p", "rb" ) )
```

```python
index_array=[]
for k in index:
    index_array.append(k)
```

```python
df_matrix= pd.DataFrame(matrix_test, index=index_array)
```

```python
df_matrix.head()
```

```python
df_matrix_transp=df_matrix.transpose()
```

```python
df_matrix_transp.head()
```

```python
dict_new=df_matrix_transp.to_dict()
```

```python
for term in dict_new.keys():
    new_list= list(dict_new[term].keys())
    
    for doc in new_list:
        if dict_new[term][doc]==0.0:
            del dict_new[term][doc]
```
